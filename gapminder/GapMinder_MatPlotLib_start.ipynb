# Plotting

Using a csv file with world development information, we are going to tell a story via plotting data.

First, let's import a couple packages to make this project easier.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the data from the gapminder.csv file
pop_minder = pd.read_csv('gapminder.csv')

# The head function outputs the first few rows.
pop_minder.head()
type(pop_minder)
```

```python
# From the pop_minder DataFrame, we create two lists.  
# One for GDP per capita and another for life expectancy.
gdp_cap = pop_minder.gdp_cap
life_exp = pop_minder.life_exp
type(pop_minder.life_exp)
```

```python
# Basic scatter plot, log scale
plt.scatter(pop_minder.gdp_cap, pop_minder.life_exp)
#plt.xscale('log')

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in Years]'
title = 'World Develop in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# After customizing, display the plot
plt.show()
```

```python
# Scatter plot
plt.scatter(gdp_cap, life_exp)

# Previous customizations
plt.xscale('log')

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

plt.xticks(tick_val, tick_lab)

plt.show()
```

```python
# Import numpy as np
import numpy as np

# Store pop as a numpy array: np_pop
pop = list(pop_minder.population/1000000)
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop * 1.5

# Update: set s argument to np_pop
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customizations
# Previous customizations
plt.xscale('log')

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.xticks(tick_val, tick_lab)


# Display the plot
plt.show()
```

```python
# Assign colors to the continents.
dict = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}

cont = list(pop_minder.cont)

col = [dict[x] for x in cont]

plt.scatter(x=gdp_cap, y=life_exp, s=np_pop, c=col, alpha=0.4)

plt.xscale('log')
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.xticks(tick_val, tick_lab)

# Show the plot
plt.show()
```

```python
# Specify c and alpha inside plt.scatter()
plt.scatter(x = gdp_cap, y = life_exp, s = np.array(pop) * 2, c = col, alpha = 0.8)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Additional customizations
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')
#plt.text(1619, )

# Add grid() call
plt.grid(True)

# Show the plot
plt.show()
```
